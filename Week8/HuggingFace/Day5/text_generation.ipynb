```python
from transformers import pipeline

generator = pipeline("text-generation",model="gpt2")

results = generator("Deepika is becoming AI Engineer because",
                    max_new_tokens=50,
                    do_sample=True,
                    temperature=0.9,
                    top_k=50,
                    top_p=0.95,
                    repetition_penalty=1.2
                    )
print(results[0]["generated_text"])
```

max_new_tokens=50

Model will generate up to 50 new words/tokens after your prompt.

If you increase this → longer output
Decrease → shorter output

🟢 do_sample=True

Very important 👇

False → Deterministic (always same output)

True → Random sampling (more creative)

You set it to True → so output can change each time.

🟢 temperature=0.9

Controls randomness.

Low (0.2–0.5) → Safe, predictable

Medium (0.7–1.0) → Balanced

High (>1.2) → Very creative, chaotic

0.9 = Good balance of creativity.

🟢 top_k=50

Model chooses next word from top 50 most probable words only.

Helps avoid very unlikely weird words.

🟢 top_p=0.95

Nucleus sampling.

Model selects words whose total probability = 95%.

This keeps generation natural.

🟢 repetition_penalty=1.2

Prevents repeating words too much.

If >1 → discourages repetition
If =1 → no penalty

1.2 = Good setting to reduce repeating sentences.
