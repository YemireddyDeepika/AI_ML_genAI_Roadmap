```python
# Temperature (Creativity Control)
# Low temperature -> safer
# High temperature -> more creative

# NOTE: model and inputs are not defined yet, this is just for demonstration of temperature control.
# You would typically define 'model' as your generative model and 'inputs' as your prompt or data.
model = None # Placeholder
inputs = {} # Placeholder

if model and inputs:
    model.generate(
        **inputs,
        max_length=20,
        temperature=0.2
    )
```

```python
## Top-K keeps only the K most probable tokens.

if model and inputs:
  model.generate(
    **inputs,
    top_k=50
    )

## Choose only from top 50 most likely words.
## Without top-k: AI future banana computer happiness science
## With top-k: AI future technology innovation research
## So it removes unlikely words.
```

```python
## Top-P (Nucleus Sampling)
## Top-P keeps the smallest set of tokens whose probabilities add up to P.

if model and inputs:
  model.generate(
      **inputs,
      top_p=0.9
  )

## Use tokens that cover 90% probability mass.
```

Example Code:

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import logging

logging.getLogger("transformers").setLevel(logging.ERROR)

tokenizer = AutoTokenizer.from_pretrained("gpt2")

tokenizer.pad_token = tokenizer.eos_token  ## it remove warnings

model = AutoModelForCausalLM.from_pretrained("gpt2")

text = "Deepika wants to become"

inputs = tokenizer(text, return_tensors="pt")
output = model.generate(**inputs,
                         max_length=30,
                         temperature=0.7,
                         top_k=50,
                         top_p=0.9,
                         do_sample=True
                         )
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

Repetition Penalty

```python
output = model.generate(
    **inputs,
    max_length=50,
    temperature=0.7,
    top_k=50,
    top_p=0.9,
    repetition_penalty=1.2,
    do_sample=True
)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```
