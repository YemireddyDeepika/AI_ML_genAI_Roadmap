```python
import torch
import torch.nn.functional as F

# Example embeddings for 3 tokens
x = torch.rand(3,4)

# Random weight matrices
wq = torch.rand(4,4)
wk = torch.rand(4,4)
wv = torch.rand(4,4)

Q = x @ wq
K = x @ wk
v = x @ wv

attention_scores = Q @ K.T
attention_weights = F.softmax(attention_scores,dim=-1)

output = attention_weights @ v

print("Attention Weights:\n", attention_weights)
print("Output:\n", output)
```

```python
import torch
import torch.nn.functional as F

x = torch.rand(4,2)

wq = torch.rand(2,3)
wk = torch.rand(2,3) # Changed from (2,2) to (2,3) for consistency with wq's output dimension
wv = torch.rand(2,3) # Changed from (3,3) to (2,3) to match x's embedding dimension (2) and wq's output dimension (3)

Q = x @ wq
K = x @ wk
V = x @ wv

attention_scores = Q @ K.T
attention_weights = F.softmax(attention_scores, dim = -1)

output = attention_weights @ V

print("Attention Weights:\n", attention_weights)
print("Output:\n", output)
```
