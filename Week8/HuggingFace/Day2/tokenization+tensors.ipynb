Step 1-Load Tokenizer and Model

```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModel.from_pretrained("gpt2")

text = "I love AI"

inputs = tokenizer(text,return_tensors="pt")

outputs = model(**inputs)

print(outputs.last_hidden_state.shape)
```

Step 2-See the Actual Vector

```python
print(outputs.last_hidden_state[0][0])
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
from transformers.utils import logging
import warnings

#Hide warnings and logs
logging.set_verbosity_error()
warnings.filterwarnings("ignore")

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModel.from_pretrained("gpt2")

text = "Deepika is learning artificial intelligence"
inputs = tokenizer(text,return_tensors="pt")
outputs = model(**inputs)

print(outputs.last_hidden_state.shape)
```

```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModel.from_pretrained("gpt2")

text = "Deepika is learning artificial intelligence"
inouts = tokenizer(text,return_tensors="pt")
outputs = model(**inouts)

#print the tokens first
tokens = tokenizer.tokenize(text)

print(tokens)

#See the vector of the first tokens
print(outputs.last_hidden_state[0][0])
```

```python
#prints the first 10 numbers of the embedding.
print(outputs.last_hidden_state[0][0][:10])
```
