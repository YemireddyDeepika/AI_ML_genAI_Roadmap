What is a Pipeline?

Pipeline = Ready-made AI tool

Instead of:

Loading model

Loading tokenizer

Processing input

Running inference

Pipeline does everything in 1 line.

It is part of the Transformers library.

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
results = classifier("I love learning AI with Hugging Face ")
print(results)
```

```python
results = classifier("I hate delays")
print(results)
```

What happened internally?

It loaded a pretrained BERT model

Loaded tokenizer

Converted text → tokens → numbers

Ran model

Returned prediction

You just used a pretrained transformer model.
