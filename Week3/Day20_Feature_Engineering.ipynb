**Feature Engineering**

1.Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

2.Dataset

```python
data = { "Study_Hours":[2,4,6,8,10,np.nan],
         "Attendance":[60,70,80,90,np.nan,75],
         "Previous_Score":[45,55,65,75,85,60],
         "Gender":["Male","Female","Female","Male","Male","Female"],
         "Pass":[0,0,1,1,1,0]}
df=pd.DataFrame(data)
df
```

3.Handling Missing Values

```python
df["Study_Hours"] = df["Study_Hours"].fillna(df["Study_Hours"].mean())
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].median())
df
```

```python
df.columns
```

4.Encoding Categorical Feature

```python
df = pd.get_dummies(df,columns=["Gender"],drop_first=True,dtype=int)
```

```python
df.columns
```

5.Creating New Features

```python
df["Study_Efficiency"] = df["Study_Hours"]/df["Attendance"]
df
```

6.Feature Scaling

```python
scaler = StandardScaler()
scaler_features = scaler.fit_transform(df[["Study_Hours","Attendance","Previous_Score","Study_Efficiency"]])
scaler_df = pd.DataFrame(scaler_features,columns=["Study_Hours","Attendance","Previous_Score","Study_Efficiency"])
scaler_df
```

7.Final Feature Set

```python
final_df = pd.concat([scaler_df,df[["Gender_Male","Pass"]]],axis=1)
final_df
```
