Step 1 --> Text Cleaning

```python
import re 

def clean_text(text):
    text = text.lower()                            # HELLO -> hello
    text = re.sub(r"http\S+", "", text)            # remove URLs
    text = re.sub(r"[^a-zA-Z\s]", " ", text)         # remove punctuation/numbers
    text = re.sub(r"\s+", " ", text).strip()        # remove extra spaces
    return text

#Text
raw = "Hello!! I LOVE AI... Visit https://ai.com it's amazing 😊"
print(clean_text(raw))
```

Step 2 --> Tokenization

```python
# ____________ Word Tokenization ________________
sentence = "I love learning AI"
tokens = sentence.split()
print(tokens)
```

```python
# ___________ Using NLTK _______________________
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

tokens = word_tokenize("I love AI! It's amazing.")
print(tokens)
```

Step 3 --> Stopword Removal

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

tokens = ["i", "love", "the", "field", "of", "ai"]
filtered = [w for w in tokens if w not in stop_words]
print(filtered)
```

Step 4 --> Stemming vs Lemmatization

```python
# ________ Stemming ________________________________
from nltk.stem import PorterStemmer
ps = PorterStemmer()

words = ["running", "studies", "happily", "caring"]
stemmed = [ps.stem(w) for w in words]
print(stemmed)
```

```python
# _________ Lemmatization _________________________________
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()

lemmatized = [lem.lemmatize(w) for w in words]
print(lemmatized)
```

Step 5 --> Building Vocabulary & Encoding

```python
# _____ Building Vocabulary ________________________
corups = [
    "i love ai",
    "ai is amazing",
    "i love deep learning"
]

# All unique words
vocab = set(word for sentence in corups for word in sentence.split())
vocab = sorted(vocab)
print("Vocab:", vocab)

# Word -> index mapping
word2idx = {word: idx+1 for idx, word in enumerate(vocab)}
print("Word2idx:", word2idx)

# ____Encode sentence ____________________________________
def encode(sentence, word2idx):
    return [word2idx[w] for w in sentence.split()]

encoded = encode("i love ai",word2idx)
print("Encoded:", encoded)
```

Step 6 -> Padding (Making equal length)

```python
from torch.nn.utils.rnn import pad_sequence
import torch

# Diffenrent length sentences encoded
s1 = torch.tensor([2,4,1])               # 3 words
s2 = torch.tensor([1,5,3,2,6])           # 5 words
s3 = torch.tensor([4,1])                 # 2 words

# Pad to same length
padded = pad_sequence([s1,s2,s3], batch_first=True)
print(padded)
```

**Full Pipeline Together**

```python
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def preprocess(text):
    # Step 1 :Clean
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    # Step 2 : Tokenize
    tokens = word_tokenize(text)

    # Step 3 : Remove Stopwords
    tokens = [t for t in tokens if t not in stop_words]

    return tokens

#Test
text = "Retrieval-Augmented Generation is an amazing NLP technique!"
print(preprocess(text))
```
