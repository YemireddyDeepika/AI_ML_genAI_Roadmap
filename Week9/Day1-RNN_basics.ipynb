**Simple RNN in pytorch** - Day1 Code

Import Libraries 

```python
import torch
import torch.nn as nn
```

Simple RNN

```python
rnn = nn.RNN(
    input_size = 4,             #Size of the each input vector
    hidden_size = 8,            #Size of the Hidden state
    num_layers = 2,             # 2 layers
    batch_first = True)

#Simulate a sentence: batch=1, sequence_length=5, input_size=4
x = torch.randn(1,5,4)

#Forword Pass
output, hidden = rnn(x)

print("Input Shape:", x.shape)
print("Output Shape:", output.shape)
print("Hidden Shape:", hidden.shape)
```
