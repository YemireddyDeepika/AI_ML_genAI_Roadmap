**RNN vs LSTM vs GRU** - Day2 Code

Import Libraries

```python
import torch
import torch.nn as nn
```

Input For all 3

```python
x = torch.randn(1,5,4)
```

**RNN**

```python
rnn = nn.RNN(input_size = 4,
             hidden_size = 8,
             batch_first = True)

output, hidden = rnn(x)

print("Input shape:", x.shape)
print("Output Shape:", output.shape)
print("Hidden Shape:", hidden.shape)
```

**LSTM**

```python
lstm = nn.LSTM( input_size = 4,
                hidden_size = 8,
                batch_first = True)

lstm_out, (lstm_h, lstm_c) = lstm(x)

print("LSTM Output Shape:", lstm_out.shape)
print("Hidden Shape:", lstm_h.shape)
print("Cell Shape:", lstm_c.shape)
```

**GRU**

```python
gru = nn.GRU(input_size = 4,
              hidden_size = 8,
              batch_first = True)

gru_out, gru_h = gru(x)

print("Output Shape:", gru_out.shape)
print("Hidden Shape:", gru_h.shape)
```
